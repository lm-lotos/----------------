25. –°—Ç–≤–æ—Ä–∏ –¥–≤—ñ –º–Ω–æ–∂–∏–Ω–∏:

`A = {1, 2, 3, 4}`  
`B = {3, 4, 5, 6}`


—ñ –∑–Ω–∞–π–¥–∏:  
- –æ–±‚Äô—î–¥–Ω–∞–Ω–Ω—è (union)
- –ø–µ—Ä–µ—Ç–∏–Ω (intersection)
- —Ä—ñ–∑–Ω–∏—Ü—é (difference)

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

# 1Ô∏è –æ–±‚Äô—î–¥–Ω–∞–Ω–Ω—è
print(A | B)              # {1, 2, 3, 4, 5, 6}
print(A.union(B))         # —Ç–µ —Å–∞–º–µ

# 2Ô∏è –ø–µ—Ä–µ—Ç–∏–Ω
print(A & B)              # {3, 4}
print(A.intersection(B))  # —Ç–µ —Å–∞–º–µ

# 3Ô∏è —Ä—ñ–∑–Ω–∏—Ü—è
print(A - B)              # {1, 2}
print(A.difference(B))    # —Ç–µ —Å–∞–º–µ

# –ø–æ—è—Å–Ω–µ–Ω–Ω—è –¥–ª—è —Å–µ–±–µ: –æ–±–∏–¥–≤–∞ —Å–ø–æ—Å–æ–±–∏ ‚Äî –∫–æ—Ä–æ—Ç–∫–∏–π –æ–ø–µ—Ä–∞—Ç–æ—Ä (|, &, -) —ñ –º–µ—Ç–æ–¥ (union(), intersection(), difference()) ‚Äî —Ä–æ–±–ª—è—Ç—å –æ–¥–Ω–µ –π —Ç–µ —Å–∞–º–µ
# A | B —ñ A.union(B) ‚Äî –æ–±‚Äô—î–¥–Ω—É—é—Ç—å –º–Ω–æ–∂–∏–Ω–∏.

# A & B —ñ A.intersection(B) ‚Äî –∑–Ω–∞—Ö–æ–¥—è—Ç—å —Å–ø—ñ–ª—å–Ω—ñ –µ–ª–µ–º–µ–Ω—Ç–∏.

# A - B —ñ A.difference(B) ‚Äî –ø–æ–∫–∞–∑—É—é—Ç—å —Ä—ñ–∑–Ω–∏—Ü—é, —Ç–æ–±—Ç–æ —â–æ —î –≤ A, –∞–ª–µ –Ω–µ–º–∞ –≤ B.

# –û–ø–µ—Ä–∞—Ç–æ—Ä–∏ –∫–æ—Ä–æ—Ç—à—ñ ‚Äî –∑—Ä—É—á–Ω—ñ—à—ñ, –∫–æ–ª–∏ –∫–æ–¥ –ø—Ä–æ—Å—Ç–∏–π.
# –ú–µ—Ç–æ–¥–∏ (.union() —Ç–æ—â–æ) ‚Äî —á–∏—Ç–∞–±–µ–ª—å–Ω—ñ—à—ñ, —è–∫—â–æ —Ç—Ä–µ–±–∞ —â–æ—Å—å —Å–∫–ª–∞–¥–Ω—ñ—à–µ –∞–±–æ –ª–∞–Ω—Ü—é–∂–∫–æ–º.
```

```python
person = {"name": "Nazar", 
          "age": 25,
          "city":"Kyiv"}  # –º–æ–∂–Ω–∞ —Å—Ç–∞–≤–∏—Ç–∏ –µ–Ω—Ç–µ—Ä –¥–ª—è –∑—Ä—É—á–Ω–æ—Å—Ç—ñ —â–æ–± –ø–∏—Å–∞—Ç–∏ –≤ —Å—Ç–æ–≤–ø—á–∏–∫
print(person["name"])
```

35. –í–∏–¥–∞–ª–∏ –∫–ª—é—á `'city'`

```python
person = {'name': 'Nazar', 'age': 26, 'city': 'Kyiv', 'country': 'Ukraine'} # –º–æ–∂–Ω–∞ —á–µ—Ä–µ–∑ "pop" –∞–±–æ —á–µ—Ä–µ–∑ "del"
 
person.pop('city')  # —Ç–æ—á–Ω—ñ—à–µ  -   person.pop('city', None)

#  –∞–±–æ  —Ç–∞–∫ -   person.del('city') 

# del person['city']

print(person)


# –ø–æ—è—Å–Ω–µ–Ω–Ω—è –¥–ª—è —Å–µ–±–µ:

# pop() –≤–∏–¥–∞–ª—è—î –∫–ª—é—á —ñ –ø–æ–≤–µ—Ä—Ç–∞—î –π–æ–≥–æ –∑–Ω–∞—á–µ–Ω–Ω—è (—è–∫—â–æ –∫–ª—é—á–∞ –Ω–µ–º–∞ ‚Äî –º–æ–∂–Ω–∞ –≤–∫–∞–∑–∞—Ç–∏ None, —â–æ–± —É–Ω–∏–∫–Ω—É—Ç–∏ –ø–æ–º–∏–ª–∫–∏).

# del –ø—Ä–æ—Å—Ç–æ –≤–∏–¥–∞–ª—è—î –∫–ª—é—á –±–µ–∑ –ø–æ–≤–µ—Ä–Ω–µ–Ω–Ω—è –∑–Ω–∞—á–µ–Ω–Ω—è.
```

36. –ü–µ—Ä–µ–≤—ñ—Ä, —á–∏ —î –∫–ª—é—á `'name'` —É —Å–ª–æ–≤–Ω–∏–∫—É

```python
# –ø—Ä–æ—Å—Ç–æ:
'name' in person

# —á–µ—Ä–µ–∑ –º–µ—Ç–æ–¥ keys():
'name' in person.keys()

# –∑ —É–º–æ–≤–æ—é:
if 'name' in person:
    print("–ö–ª—é—á 'name' —î –≤ —Å–ª–æ–≤–Ω–∏–∫—É")

# –æ–ø–µ—Ä–∞—Ç–æ—Ä in –ø–µ—Ä–µ–≤—ñ—Ä—è—î, —á–∏ —ñ—Å–Ω—É—î –∫–ª—é—á —É —Å–ª–æ–≤–Ω–∏–∫—É (–Ω–µ –∑–Ω–∞—á–µ–Ω–Ω—è!).
# –¢–æ–º—É in person ‚Äî —Ü–µ –Ω–∞–π–∫–æ—Ä–æ—Ç—à–∏–π —ñ –ø—Ä–∞–≤–∏–ª—å–Ω–∏–π –≤–∞—Ä—ñ–∞–Ω—Ç.
```

37. –ü—Ä–æ–π–¥–∏ —Ü–∏–∫–ª–æ–º –ø–æ —Å–ª–æ–≤–Ω–∏–∫—É —ñ –≤–∏–≤–µ–¥–∏ –≤—Å—ñ –∫–ª—é—á—ñ —Ç–∞ –∑–Ω–∞—á–µ–Ω–Ω—è.

```python
# –æ–±–∏—á–Ω–∏–π —Ü–∏–∫–ª:
for key, value in person.items():
    print(key, value)

# –æ–∫—Ä–µ–º–æ –∫–ª—é—á—ñ:
for key in person:
    print(key, person[key])

# –∞–±–æ –∫—Ä–∞—â–µ —ñ –∫–æ—Ä–æ—Ç—à–µ:
for k, v in person.items():
    print(f"{k}: {v}")

#  .items() –ø–æ–≤–µ—Ä—Ç–∞—î –ø–∞—Ä–∏ (–∫–ª—é—á, –∑–Ω–∞—á–µ–Ω–Ω—è),
# —Ç–æ–º—É –∑—Ä—É—á–Ω–æ —Ä–æ–∑–ø–∞–∫–æ–≤—É–≤–∞—Ç–∏ —ó—Ö –æ–¥—Ä–∞–∑—É —É –¥–≤—ñ –∑–º—ñ–Ω–Ω—ñ.
```
