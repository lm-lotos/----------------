````python
# Розв'язання для завдань 27–35
# Підказка: можеш згенерувати випадкові дані за допомогою np.random.

# Твій код тут ↓
````

````python
# 27. Створюємо DataFrame з даними про продажі

np.random.seed(1)

sales = pd.DataFrame({
    'Product': np.random.choice(['Phone', 'Laptop', 'Tablet'], 12),
    'Category': np.random.choice(['Electronics', 'Accessories'], 12),
    'Price': np.random.randint(100, 1000, 12),
    'Quantity': np.random.randint(1, 10, 12)
    })
print(sales)
````

````python
# 28. Додаємо стовпець із загальним доходом
sales['Revenue'] = sales['Price'] * sales['Quantity']
sales   
# sales.head()  пишемо, якщо велика таблиця
#  і, відповідно в пайтоні: print(sales), або, якощо велика таблиця - то print(sales.head())
# перевірка: sales.columns 
# або ще краще: sales = sales.assign(Revenue=sales['Price'] * sales['Quantity'])
````

````python
# 29. Обчислюємо загальний дохід по кожній категорії
sales.groupby('Category')['Revenue'].sum()
````

````python
# 30. Обчислюємо середню ціну для кожного товару
sales.groupby('Product')['Price'].mean()
````

````python
# 31. Визначаємо найпопулярніший товар за кількістю продажів
sales.groupby('Product')['Quantity'].sum().idxmax()
````

````python
# 32. Знаходимо мінімальний та максимальний дохід по категоріях
sales.groupby('Category')['Revenue'].agg(['min', 'max'])
````

````python
# 33. Створюємо зведену таблицю із середнім доходом по категоріях
pivot = pd.pivot_table(
    sales,
    index='Category',
    values='Revenue',
    aggfunc='mean'
)
print(pivot)

# читаємо назви колонок і округлюємо, щоб не було довгих хвостів:
# pivot = sales.groupby('Category')['Revenue'].mean().round(2)
# можна і без pivot_table, тоді через groupby
````

````python
# 34. Сортуємо категорії за середнім доходом
pivot_sorted = pivot.sort_values(by='Revenue', ascending=False)
print(pivot_sorted)
# щоб побачити результат - сортуємо в зворотньому порядку - від меншого до більшого
````

````python
# 35. Додаємо стовпець з прибутком і знаходимо середній прибуток по категоріях
sales['Profit'] = sales['Revenue'] * 0.25
sales.groupby('Category')['Profit'].mean()
````

````python
# 61. Створюємо DataFrame з даними про продажі
sales_full = pd.DataFrame({
    'date': pd.to_datetime([
        '2024-01-01', '2024-01-05', '2024-01-10',
        '2024-02-03', '2024-02-10', '2024-02-20',
        '2024-03-05', '2024-03-15', '2024-03-25'
    ]),
    'product': ['Phone', 'Laptop', 'Tablet',
                'Phone', 'Laptop', 'Tablet',
                'Phone', 'Laptop', 'Tablet'],
    'category': ['Electronics'] * 9,
    'price': [500, 1000, 700, 520, 980, 710, 510, 1020, 690],
    'quantity': [2, 1, 3, 1, 2, 1, 3, 1, 2],
    'city': ['Kyiv', 'Lviv', 'Kyiv',
             'Odesa', 'Kyiv', 'Lviv',
             'Kyiv', 'Odesa', 'Lviv']
})

# додаємо дохід
sales_full['Revenue'] = sales_full['price'] * sales_full['quantity']

# показуємо таблицю
sales_full
````

````python
# 62. Рахуємо загальний дохід по місяцях
monthly_revenue = (
    sales_full
    .set_index('date')
    .resample('M')['Revenue']
    .sum()
)

# дивимось результат
monthly_revenue
````

````python
# будуємо лінійний графік доходу по місяцях
monthly_revenue.plot(
    marker='o',
    linewidth=2,
    title='Загальний дохід по місяцях'
)
plt.xlabel('Місяць')
plt.ylabel('Дохід')
plt.grid(True)
plt.show()
````

````python
# 63. Топ-3 товари за доходом + стовпчиковий графік
top_products = (
    sales_full.groupby('product')['Revenue']
    .sum()
    .sort_values(ascending=False)
    .head(3)
)

top_products

top_products.plot(kind='bar', title='Топ-3 товари за доходом')
plt.show()
````

````python
# створюємо DataFrame
data = {
    'city': ['Kyiv', 'Lviv', 'Odesa'],
    'revenue': [4200, 3100, 3600]
}

df = pd.DataFrame(data)

# будуємо графік
plt.figure(figsize=(8, 5))
plt.bar(df['city'], df['revenue'])

plt.title('Revenue by city')
plt.xlabel('City')
plt.ylabel('Revenue')

plt.show()
````

````python
# 64. Рахуємо загальний дохід по кожному місту
city_revenue = sales_full.groupby('city')['Revenue'].sum()

# показуємо таблицю з доходами по містах
city_revenue
````

````python
# будуємо стовпчиковий графік доходів по містах
city_revenue.plot(
    kind='bar',
    title='Дохід по містах'
)

plt.xlabel('Місто')
plt.ylabel('Дохід')
plt.show()
````

````python
# 65. Обчислюємо кореляцію між числовими стовпцями
corr = sales_full[['price', 'quantity', 'Revenue']].corr()

# показуємо таблицю кореляції
corr
````

````python
# будуємо heatmap кореляції
plt.imshow(corr)
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title('Кореляція між числовими змінними')
plt.show()
````

````python
# 66. Створюємо subplot з двома графіками

fig, ax = plt.subplots(1, 2, figsize=(10, 4))

# ліворуч — лінійний графік доходу по місяцях
monthly_revenue.plot(
    ax=ax[0],
    marker='o',
    linewidth=2,
    title='Дохід по місяцях'
)
ax[0].set_xlabel('Місяць')
ax[0].set_ylabel('Дохід')
ax[0].grid(True, alpha=0.3)
# праворуч — стовпчиковий графік доходу по містах
city_revenue.plot(
    kind='bar',
    ax=ax[1],
    title='Дохід по містах'
)
ax[1].set_xlabel('Місто')
ax[1].set_ylabel('Дохід')

plt.tight_layout()
plt.show()

````

````python
# 67. Порівняння доходів за два роки (без легенди)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
revenue_2023 = [3000, 3200, 3100, 3300, 3400, 3500]
revenue_2024 = [4000, 3500, 3200, 3700, 2900, 3100]
plt.plot(months, revenue_2024, marker='o')
plt.xlabel('Місяць')
plt.ylabel('Дохід')
plt.show()
````

````python
# 68. Дохід (стовпчики) + середня ціна (лінія)
avg_price = sales_full.groupby('month')['price'].mean()

# показуємо таблицю
avg_price
````

````python
# Графік з двома осями: дохід + середня ціна

fig, ax1 = plt.subplots()

# стовпчики — дохід
ax1.bar(monthly_revenue.index.astype(str), monthly_revenue.values)
ax1.set_xlabel('Місяць')
ax1.set_ylabel('Дохід')

# друга вісь — середня ціна
ax2 = ax1.twinx()
ax2.plot(avg_price.index.astype(str), avg_price.values)
ax2.set_ylabel('Середня ціна')

plt.title('Дохід та середня ціна по місяцях')
plt.show()
````
